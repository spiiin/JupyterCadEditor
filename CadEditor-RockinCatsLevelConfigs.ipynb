Скрипт вывода информации об адресах блоков для Rocking Cats

<pre>minor_blocks - блок из 2х2 тайла + 1 байт id палитры
major_blocks_ptrs - блок из 4х1 minor_block
maj_major_blocks_ptrs - блок из 1х4 major_block
maj_maj_major_blocks_ptrs - блок из 2х1 maj_major_block
room - блок из 2x3 maj_maj_major_block

room_ptrs: 0x8eb9 + 0xa010 = 0x12EC9
minor_blocks_ptrs: 0x9091 + 0xa010 = 0x130A1;
major_blocks_ptrs: 0x901b + 0xa010 = 0x1302B;
maj_major_blocks_ptrs: 0x8fa5 + 0xa010 = 0x12FB5;
maj_maj_major_blocks_ptrs: 0x8f2f + 0xa010 = 0x12F3F;</pre>

```python
romname = r"d:\DEV\CadEditor\CadEditor\Rockin' Kats (U) [!].nes"
with open(romname, "rb") as f:
    d = f.read()
    d = map(lambda b: ord(b), d)
```

```python
def readWord(d, addr):
    return d[addr] + (d[addr+1]<<8)

for actNo in xrange(59):
    print "--------------------------------"
    print "LEVEL", actNo
    b = readWord(d, 0x130A1 + actNo*2)
    bb = readWord(d, 0x1302B + actNo*2)
    bbb = readWord(d, 0x12FB5 + actNo*2)
    bbbb = readWord(d, 0x12F3F + actNo*2)
    s = readWord(d, 0x12EC9 + actNo*2)
    print "Blocks ptr          :",  hex(b), "->",hex(b + 0xa010), "count:", (bb-b)/5.0
    print "Big blocks ptr      :", hex(bb), "->",hex(bb + 0xa010), "count:", (bbb-bb)/4.0
    print "BigBig blocks ptr   :", hex(bbb), "->",hex(bbb + 0xa010), "count:", (bbbb-bbb)/4.0
    print "BigBigBig blocks ptr:", hex(bbbb), "->",hex(bbbb + 0xa010), "count:", (s-bbbb)/2.0
    print "Screens             :", hex(s), "->",hex(s + 0xa010)
    
    v = d[0x12519 + actNo]
    print "Video addr          :", hex(v), "->", hex(d[0x128bc+v*2]),hex(d[0x128bc+v*2+1])
```
